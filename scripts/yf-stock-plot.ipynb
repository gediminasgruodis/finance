```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

# Set style
plt.style.use('ggplot')

# Define the ticker symbols
ticker1 = 'TSLA'
ticker2 = 'MSFT'

# Download data using yfinance
data1 = yf.download(ticker1)['Adj Close']
data2 = yf.download(ticker2)['Adj Close']

# Create a figure and axis objects
fig, ax = plt.subplots(figsize=(18, 6))

# Plot the data
ax.plot(data1.index, data1, label=ticker1, color='blue')
ax.plot(data2.index, data2, label=ticker2, color='green')

# Set labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Adjusted Close Price')
ax.set_title(f'{ticker1} vs. {ticker2}')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
