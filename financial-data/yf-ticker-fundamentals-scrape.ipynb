```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Enter a stock symbol
index = 'MSFT'

# URL links
url_is = f'https://finance.yahoo.com/quote/{index}/financials?p={index}'
url_bs = f'https://finance.yahoo.com/quote/{index}/balance-sheet?p={index}'
url_cf = f'https://finance.yahoo.com/quote/{index}/cash-flow?p={index}'

    # User-Agent header
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}

# Send requests with headers
response_is = requests.get(url_is, headers=headers)
response_bs = requests.get(url_bs, headers=headers)
response_cf = requests.get(url_cf, headers=headers)

# Check responses
if response_is.status_code == 200:
    print("Financials data retrieved successfully.")
    soup_is = BeautifulSoup(response_is.content, 'html.parser')  # Corrected to 'html.parser'
else:
    print(f"Failed to retrieve financials data. Status Code: {response_is.status_code}")
```

```python
ls = []  # Create an empty list to store data
for l in soup_is.find_all('div'):  # Find all 'div' elements in the BeautifulSoup object
    ls.append(l.string)  # Add each element's text content to the list

# Create a new list 'ls' excluding elements with text content 'Operating Expenses' and 'Non-recurring Events'
ls = [e for e in ls if e not in ('Operating Expense', 'Non-recurring Events')]

# Filter out the 'None' values
ls = list(filter(None,ls))

# Select only relevant items
ls = ls[15:]

# Fix the missing 'Total Revenue' column name
# Find the index of '6/30/2020'
index_2020 = ls.index('6/30/2020')

# Insert 'Total Revenue' at the index next to '6/30/2020'
ls.insert(index_2020 + 1, 'Total Revenue')

# Convert to a list
is_data = list(zip(*[iter(ls)]*5))
ls
#is_data
```

```python
# Add everything to a dataframe
income_st = pd.DataFrame(is_data[0:])
income_st
```

```python
Income_st.columns = Income_st.iloc[0]  # Name columns to the first row of the dataframe
Income_st = Income_st.iloc[1:]  # Start reading from the 1st row
Income_st = Income_st.T  # Transpose the dataframe
Income_st.columns = Income_st.iloc[0]  # Name columns to the first row of the dataframe
Income_st.drop(Income_st.index[0], inplace=True)  # Drop the first index row
Income_st.index.name = ''  # Remove the index name
Income_st.rename(index={'ttm': '12/31/2019'}, inplace=True)  # Rename 'ttm' in the index columns to '12/31/2019'
Income_st = Income_st.iloc[:, :-5]  # Remove the last 5 irrelevant columns
```
